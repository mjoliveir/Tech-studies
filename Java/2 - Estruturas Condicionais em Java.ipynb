Estruturas condicionais permitem controlar o fluxo do programa com base em decisões lógicas. Em Java, existem três formas principais:

## if / else if / else

Permite executar blocos de código diferentes com base em uma ou mais condições booleanas.

```java
if (idade >= 18) {
    System.out.println("Maior de idade");
} else if (idade > 0) {
    System.out.println("Menorteste de idade");
} else {
    System.out.println("Idade inválida");
}
```

## Operador ternário

Forma compacta de uma estrutura `if/else` para atribuições simples.

```java
String status = (idade >= 18) ? "Maior" : "Menor";
```

## [[./Switch-case|Switch-case]]

Mais apropriado quando se deseja comparar um valor específico com várias possibilidades.

```java
int dia = 3;
switch (dia) {
    case 1:
        System.out.println("Domingo");
        break;
    case 2:
        System.out.println("Segunda-feira");
        break;
    case 3:
        System.out.println("Terça-feira");
        break;
    default:
        System.out.println("Dia inválido");
}
```

## Boas Práticas

- Use `if/else` para decisões com condições complexas ou booleanas.
    
- Prefira `switch` para verificações de múltiplos valores fixos (como enums ou inteiros).
    
- Evite aninhamentos profundos que dificultam a leitura.
    
- Sempre inclua um `default` em estruturas `switch`.
