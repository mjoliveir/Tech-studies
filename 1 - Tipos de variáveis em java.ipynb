Em Java, variáveis são usadas para armazenar dados que podem ser manipulados ao longo da execução de um programa. Elas são fortemente tipadas, o que significa que cada variável precisa ser declarada com um tipo específico.

## Classificação Geral

### 1. Variáveis Primitivas

São os tipos básicos de dados. Não são objetos e armazenam valores simples.

|Tipo|Descrição|Tamanho|Exemplo|
|---|---|---|---|
|byte|Números inteiros pequenos|8 bits|byte idade = 20;|
|short|Inteiros curtos|16 bits|short ano = 2024;|
|int|Inteiros padrão|32 bits|int numero = 1000;|
|long|Inteiros longos|64 bits|long populacao = 8000000000L;|
|float|Números decimais com menor precisão|32 bits|float preco = 19.99f;|
|double|Decimais com maior precisão|64 bits|double saldo = 1500.75;|
|char|Caractere único|16 bits|char letra = 'A';|
|boolean|Verdadeiro ou falso|1 bit|boolean ativo = true;|

### 2. Variáveis de Referência

Armazenam referências para objetos, não os dados propriamente ditos.

Exemplos:

```java
String nome = "João";
int[] numeros = {1, 2, 3};
MinhaClasse objeto = new MinhaClasse();
```

### 3. Variáveis Locais

Declaradas dentro de métodos. Precisam ser inicializadas antes do uso.

### 4. Variáveis de Instância

Declaradas dentro da classe, mas fora de qualquer método. Pertencem a uma instância da classe (objeto).

### 5. Variáveis de Classe (estáticas)

Declaradas com o modificador `static`. Pertencem à classe, e não às instâncias.

```java
public class Exemplo {
    static int contador = 0; // Variável de classe
    int idade;               // Variável de instância

    public void metodo() {
        String nome = "Lucas"; // Variável local
    }
}
```

## Boas Práticas

- Sempre escolha o tipo mais adequado ao dado.
    
- Nomeie variáveis de forma descritiva.
    
- Inicialize variáveis antes de usá-las, principalmente as locais.
